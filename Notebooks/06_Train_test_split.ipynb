```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("../Dataset/preprocessed_dataset.csv")
print("Loaded shape:", df.shape)

X = df.drop(columns=["Is Delayed"])
y = df["Is Delayed"]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% for testing, 80% for training
    random_state=42,    # fixes the "shuffle" so results are reproducible every time
    stratify=y          # keeps the same delay/no-delay ratio in both train and test
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
print("Delay rate in full data:", y.mean().round(4))
print("Delay rate in y_train:", y_train.mean().round(4))
print("Delay rate in y_test:", y_test.mean().round(4))
```

```python
X_train.to_csv("../Dataset/x_train.csv", index=False)
X_test.to_csv("../Dataset/x_test.csv", index=False)
y_train.to_csv("../Dataset/y_train.csv", index=False)
y_test.to_csv("../Dataset/y_test.csv", index=False)

print("Train/test files saved successfully!")
print("X_train:", X_train.shape, "| X_test:", X_test.shape)
print("y_train:", y_train.shape, "| y_test:", y_test.shape)
```
