```python

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("C:/Users/CHENREDDY BHAVYA/SupplyPrescript/Dataset/Preprocessed_dataset.csv")
df.head()
```

```python
print("Dataset Shape:",df.shape)
print("\nColumns:")
print(df.columns)
print("\nData Types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())

```

```python
print(df.columns.tolist())
```

```python
y = df['Scheduled Delivery Date']
x = df.drop('Scheduled Delivery Date',axis=1)
```

```python
print("Feature shape:",x.shape)
print("Target shape:",y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split( x,y,test_size=0.2,random_state=42)
   
```

```python

print("Training Features:",x_train.shape)
print("Testing Features:",x_test.shape)
print("Training Labels:",y_train.shape)
print("Testing Labels:",y_test.shape)
```

```python
x_train.to_csv("../Dataset/x_train.csv", index=False)
x_test.to_csv("../Dataset/x_test.csv", index=False)

y_train.to_frame(name="Scheduled Delivery Date").to_csv("../Dataset/y_train.csv", index=False)
y_test.to_frame(name="Scheduled Delivery Date").to_csv("../Dataset/y_test.csv", index=False)

print("Train and test datasets saved successfully!")
```
